```python
#find  columns with missing value :
missing_counts = files_sold.isna().sum()
print(missing_counts)
```

```python
#missing percentage per columns:
missing_percentages = (files_sold.isna().mean() * 100).round(2)
print(missing_percentages)
```

```python
#Flag columns with >90% missing values
missing_flag = missing_percentages[missing_percentages > 90].index

print(missing_flag)
```

```python
new_data = files_sold.drop(columns=missing_flag)
new_data
```
