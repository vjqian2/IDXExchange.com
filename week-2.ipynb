```python
# Validate completeness
files_sold.isnull().sum()
```

```python
#find data type for sold and list
print(files_sold.info())
```

```python
#find  columns with missing value :
missing_counts_sold = files_sold.isna().sum()
print(files_sold.info())

print(missing_counts_sold)
```

```python
#missing percentage per columns for sold:
missing_percentages = (files_sold.isna().mean() * 100).round(2)
print(missing_percentages)
```

```python
#Flag columns with >90% missing values
missing_flag = missing_percentages[missing_percentages > 90].index

print(missing_flag)
```

```python
new_data_sold = files_sold.drop(columns=missing_flag)
new_data_sold
```

```python
print(new_data_sold.shape)
```

```python
#doing the same thing for listing file
print(files_list.info())
```

```python
missing_counts_list = files_list.isna().sum()
print(files_list.info())

print(missing_counts_list)
```

```python
#missing percentage per columns for sold:
missing_percentages_list = (files_list.isna().mean() * 100).round(2)
print(missing_percentages_list)
```

```python
missing_flag_list = missing_percentages_list[missing_percentages_list > 90].index

print(missing_flag_list)
```

```python
new_data_list = files_list.drop(columns=missing_flag_list)
new_data_list
```

```python
#save new filtered data as csv
new_data_sold.to_csv("W2_sold_data.csv", index=False)
new_data_list.to_csv("W2_list_data.csv", index=False)
```
