```python
fields_list = pd.read_csv("W3_list_enriched.csv")
fields_list
fields_sold = pd.read_csv("W3_sold_enriched.csv")
fields_sold
```

## Drop redundant and unnecessary data

```python
#drop redudent and unnessary data
Columns_drop_sold = [
    'BuyerAgencyCompensation',
    'BuyerAgencyCompensationType',
    'OriginatingSystemName',
    'PropertyType'
]
fields_sold.columns
fields_sold.drop(columns=Columns_drop_sold, inplace=True)
fields_sold
```

```python
fields_list.drop(columns=Columns_drop_list, inplace=True)
fields_list
```

```python
#Handle missing values appropriately
missing_sold = (fields_sold.isna().sum()).sort_values(ascending=False)
missing_list = (fields_list.isna().sum()).sort_values(ascending=False)
print(missing_sold)
print(missing_list)
```

```python
#fill numeric NaN with medium to ensure the dataset is filled
numeric_cols_sold = fields_sold.select_dtypes(include='number').columns
fields_sold[numeric_cols_sold] = fields_sold[numeric_cols_sold].fillna(
    fields_sold[numeric_cols_sold].median()
)

numeric_cols_list = fields_list.select_dtypes(include='number').columns
fields_list[numeric_cols_list] = fields_list[numeric_cols_list].fillna(
    fields_list[numeric_cols_list].median()
)
```

```python
missing_sold = (fields_sold.isna().sum()).sort_values(ascending=False)
print(missing_sold)
```

```python
#Ensure numeric fields are properly typed
numeric_fields = ["ClosePrice", "ListPrice", "OriginalListPrice", "LivingArea","LotSizeAcres", "BedroomsTotal", "BathroomsTotalInteger", "DaysOnMarket", "YearBuilt"]
for col in numeric_fields:
    fields_sold[col] = pd.to_numeric(fields_sold[col], errors='coerce')
print(fields_sold[numeric_fields].dtypes)
```

## Flag invalid numeric values 

```python
#flag invalid numeric values for sold: 
# ClosePrice <= 0, LivingArea <= 0, DaysOnMarket < 0,negative Bedrooms or Bathrooms
fields_sold["InvalidValues"]= (
    (fields_sold["ClosePrice"] <= 0) |
    (fields_sold["LivingArea"] <= 0) |
    (fields_sold["DaysOnMarket"] < 0) |
    (fields_sold["BedroomsTotal"] < 0) |
    (fields_sold["BathroomsTotalInteger"] < 0)
)

fields_sold["InvalidValues"]
fields_sold["InvalidValues"].value_counts()
```

```python
#finding the clomums that are invalid
fields_sold[fields_sold["InvalidValues"] == True]
```

```python
#do the same process for list: 
fields_list["InvalidValues"]= (
    (fields_list["ClosePrice"] <= 0) |
    (fields_list["LivingArea"] <= 0) |
    (fields_list["DaysOnMarket"] < 0) |
    (fields_list["BedroomsTotal"] < 0) |
    (fields_list["BathroomsTotalInteger"] < 0)
)

fields_list["InvalidValues"]
fields_list["InvalidValues"].value_counts()
```

```python
fields_list[fields_list["InvalidValues"] == True]
```

# Data consistency checks for sold and list

```python
#data consistency checks for sold
fields_sold["listing_after_close_flag"] = fields_sold["ListingContractDate"] > fields_sold["CloseDate"]
fields_sold["purchase_after_close_flag"] = fields_sold["PurchaseContractDate"] > fields_sold["CloseDate"]
fields_sold["negative_timeline_flag"] = ((fields_sold["PurchaseContractDate"] < fields_sold["ListingContractDate"]) |
    (fields_sold["CloseDate"] < fields_sold["PurchaseContractDate"]))
fields_sold
```

```python
fields_sold["purchase_after_close_flag"].value_counts()
```

```python
#data consistency checks for list
fields_list["listing_after_close_flag"] = fields_list["ListingContractDate"] > fields_list["CloseDate"]
fields_list["purchase_after_close_flag"] = fields_list["PurchaseContractDate"] > fields_list["CloseDate"]
fields_list["negative_timeline_flag"] = ((fields_list["PurchaseContractDate"] < fields_list["ListingContractDate"]) |
    (fields_list["CloseDate"] < fields_list["PurchaseContractDate"]))
fields_list
```

## Geographic Data Checks

```python
#Flag records with missing coordinates (Latitude or Longitude is null)
fields_sold["missing_sold"] = fields_sold["Latitude"].isna() | fields_sold["Longitude"].isna()
fields_sold
fields_sold["missing_sold"].value_counts()
```

Result: There is no null or missing data for longitude or latitude

```python
#Flag Latitude = 0 or Longitude = 0 (sentinel null values)
fields_sold["sentinel_null_values"] = ((fields_sold["Latitude"] == 0) | (fields_sold["Longitude"] == 0))
fields_sold
```

```python
fields_sold["positive_logitude_error"] = (fields_sold["Longitude"] > 0 )
fields_sold
fields_sold["positive_logitude_error"].value_counts()
```

There are 29 positive longitude error

```python
# Flag out-of-state or implausible coordinates
fields_sold["Invalidcoordinates"] = (
    (fields_sold["Latitude"] < -90) |
    (fields_sold["Latitude"] > 90) |
    (fields_sold["Longitude"] < -180) |
    (fields_sold["Longitude"] > 180) | (fields_sold["Latitude"] < 32.5) |
    (fields_sold["Latitude"] > 42.0) |
    (fields_sold["Longitude"] < -124.5) |
    (fields_sold["Longitude"] > -114.0))
fields_sold["Invalidcoordinates"].value_counts()
```

```python
fields_sold[fields_sold["Invalidcoordinates"] == True]
```

Based on the result, there are 88 data that have invalid coordinates. The coordinate for california is 32.5-42 for latitude and -124.5 to -114.0 for longitute, anything outside of the range is considered outside of state. Additionally, a laitiude must be between -90 and 90 and longitude must be between -180 to 180. Based on there data there are some out of state and impossible coordinates.
